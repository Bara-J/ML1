## Hřebenová regrese

Hřebenová regrese může být vhodným modelem pro regresní úlohu, její výhodou je navíc to, že nemá problém s kolinearitou.

````python
from sklearn.linear_model import Ridge, LinearRegression
````

Při použití hřebenové regrese bývá vhodné data nějakým způsobem normalizovat, protože tento model je ovlivňován škálou dat. Vyzkouším ted min-max normalizaci a standardizaci.

### Normalizace a hřebenová regrese

````python
Xtrain, Xval, Xtest, ytrain, yval, ytest

from sklearn.preprocessing import MinMaxScaler

# Standardizace trénovacích a validačních dat
scaler = MinMaxScaler()
Xtrain_scaled1 = scaler.fit_transform(Xtrain)
Xval_scaled1 = scaler.transform(Xval)

# Trénování Ridge modelu na škálovaných datech
clf_ridge_scaled1 = Ridge(1)
clf_ridge_scaled1.fit(Xtrain_scaled1, ytrain)


print(f'koeficienty: {clf_ridge_scaled1.coef_}')
print(f'Intercept: {clf_ridge_scaled1.intercept_:.4f}')

RMSE_val_ridge1 = np.sqrt(mean_squared_error(yval, clf_ridge_scaled1.predict(Xval_scaled1)))
print(f'Validační RMSE: {RMSE_val_ridge1:.5f}')


plt.scatter(yval, clf_ridge_scaled1.predict(Xval_scaled1))
plt.plot([20,90], [20,90], 'r')
plt.show()
````

Koeficienty jsou stále "rozlítané", zkusím standardizaci...

### Standardizace a hřebenová regrese

````python
from sklearn.preprocessing import StandardScaler

# Standardizace trénovacích a validačních dat
scaler = StandardScaler()
Xtrain_scaled2 = scaler.fit_transform(Xtrain)
Xval_scaled2 = scaler.transform(Xval)
Xtest_scaled2 = scaler.transform(Xtest)

# Trénování Ridge modelu na škálovaných datech
clf_ridge_scaled2 = Ridge(1)
clf_ridge_scaled2.fit(Xtrain_scaled2, ytrain)


print(f'koeficienty: {clf_ridge_scaled2.coef_}')
print(f'Intercept: {clf_ridge_scaled2.intercept_:.4f}')

RMSE_val_ridge2 = np.sqrt(mean_squared_error(yval, clf_ridge_scaled2.predict(Xval_scaled2)))
print(f'Validační RMSE: {RMSE_val_ridge2:.5f}')


plt.scatter(yval, clf_ridge_scaled2.predict(Xval_scaled2))
plt.plot([0,90], [0,90], 'r')
plt.show()
````

Koeficienty jsou všechny "podobně malé", standardizace bude vhodnější pro tento dataset, použiji ji proto pro transformaci dat před provedením hřebenové regrese, pro kterou následně vyladím hyperparametr alpha (=lambda).

````python
from scipy import optimize

def get_opt_ridge_model(Xtrain_scaled2, Xval_scaled2):
    def ridgemodel_eval(alpha):
        clf = Ridge(alpha=alpha)
        clf.fit(Xtrain_scaled2, ytrain)
        return np.sqrt(mean_squared_error(yval, clf.predict(Xval_scaled2)))

    alphas = np.linspace(0.001,0.2,100)
    alphas_res = [ridgemodel_eval(alpha) for alpha in alphas]
    plt.plot(alphas, alphas_res, '.')
    plt.show()
            
    opt_alpha = optimize.minimize_scalar(ridgemodel_eval, options = {'maxiter': 50}, method = 'bounded', bounds=(0.1, 400))
    print('Optimal alpha', opt_alpha)
    print("\n")

    clf_opt_ridge = Ridge(alpha = opt_alpha.x)
    clf_opt_ridge.fit(Xtrain_scaled2, ytrain)
    return clf_opt_ridge

clf_opt_ridge = get_opt_ridge_model(Xtrain_scaled2, Xval_scaled2)
RMSE_val_opt_ridge = np.sqrt(mean_squared_error(yval, clf_opt_ridge.predict(Xval_scaled2)))
print(f'Validační RMSE: {RMSE_val_opt_ridge:.5f}')

plt.plot(clf_opt_ridge.coef_, 'b+', label="Optimal ridge regression")
plt.legend()
plt.show()
````

````python
from sklearn.metrics import mean_absolute_error
MAE_val_opt_ridge = mean_absolute_error(yval, clf_opt_ridge.predict(Xval_scaled2)) 
print(f'Validační RMSE: {RMSE_val_opt_ridge:.5f}')
print(f'Validační MAE: {MAE_val_opt_ridge:.5f}')
````

Optimální alpha (=lambda) byla určena v hodnotě 0.10000, pro tento optimalizovaný modelu pak vychází RMSE na validační množině 2.07726 a MAE 1.29328.

### Ladění hyperparametrů

Jako hyperparamtery, které budu ladit, jsem vybrala maximální hloubku stromu (v rozmezí 1-30) a kritérium pro měření kvality rozdělení, kde se použije buď squared_error nebo absolute_error.

````python
from sklearn.model_selection import ParameterGrid
param_grid = {
    'max_depth': range(1,30), 
    'criterion': ['squared_error', 'absolute_error'],
    'min_samples_leaf': range(2,10)
}
param_comb = ParameterGrid(param_grid)
````

Naučím model pro všechny hyperparametry pomocí metody `.fit()` a naměřím přesnost na trénovacích a validačních datech.

````python
from tqdm import tqdm
val_acc = []
train_acc = []
for params in tqdm(param_comb):
    dt = DecisionTreeRegressor(max_depth=params['max_depth'], criterion=params['criterion'], min_samples_leaf=params['min_samples_leaf'], random_state = 42)
    dt.fit(Xtrain, ytrain)
    train_acc.append(np.sqrt(mean_squared_error(ytrain, dt.predict(Xtrain))))
    val_acc.append(np.sqrt(mean_squared_error(yval, dt.predict(Xval))))
````

Provedu grafické zobrazení RMSE na trénovací a validační množině.

````python
plt.figure(figsize=(15,5))
plt.plot(train_acc,'or-')
plt.plot(val_acc,'ob-')
plt.xlabel('hyperparametr index')
plt.ylabel('RMSE')
plt.legend(['train', 'validation'])
````

Lze vidět, že na trénovací množině se RMSE s rostoucí hloubkou stromu dostane až na hodnotu blízkou 0 (stromy jsou přeučené). Musím zvolit nejvhodnější kombinaci hyperparametrů (hledám tedy nejnižší hodnotu RMSE na validačních datech).

````python
# Najdu nejvhodnější parametry
best_params = param_comb[np.argmin(val_acc)]
print('best params', best_params)
````

Jako nejlepší kombinace hyperparametrů byla určena maximální hloubka stromu 26, absolute_error jako criterion a minimální počet bodů v listu 2.
Nyní tyto hyperparametry použiji pro naučení rozhodovacího stromu.

````python
dt = DecisionTreeRegressor(max_depth=best_params['max_depth'], criterion=best_params['criterion'], min_samples_leaf=best_params['min_samples_leaf'], random_state = 42)
dt.fit(Xtrain, ytrain)

# Vypsání přesnosti na trénovací a validační množině
print('RMSE (train): {0:.5f}'.format(np.sqrt(mean_squared_error(ytrain, dt.predict(Xtrain)))))
print('RMSE (val): {0:.5f}'.format(np.sqrt(mean_squared_error(yval, dt.predict(Xval)))))
print('MAE (val): {0:.5f}'.format(np.sqrt(mean_absolute_error(yval, dt.predict(Xval)))))
RMSE_dt = (np.sqrt(mean_squared_error(yval, dt.predict(Xval))))
MAE_dt = mean_absolute_error(yval, dt.predict(Xval)) 
````

## Random Forest

Náhodný les patří mezi tzv. ensembl metody, které se skládají z více podmodelů a výsledek skládají z jejich rozhodnutí. Pro tuto úlohu by také mohl být vhodným kandidátem, jelikož se pomocí něho dají predikovat spojité proměnné (tato úloha je právě úlohou regresní). Náhodné lesy jsou na rozdíl od rozhodovacích stromů velice robustní a poměrně
odolné vůči přeučení, nevýhodou ale je, že částečně ztrácejí jejich jednoduchost a snadnou
interpretovatelnost.

````python
# Předpřipravená kostra modelu náhodného lesa
class CustomRandomForest:
    """
    Třída Vašeho modelu
    Bude se jednat o model náhodného lesa, kde podmodely tvoří rozhodovací stromy pro regresi.
    Pro podmodely můžete použít implementaci DecisionTreeRegressor ze sklearn.
    """
    def __init__(self, n_estimators, max_samples, max_depth, **kwargs):
        """
        Konstruktor modelu
        Základní hyperparametery:
            n_estimators - počet podmodelů - rozhodovacích stromů.
            max_samples - vyberte si, zda tento parametr bude označovat relativní počet bodů (tj. číslo mezi 0 a 1) 
                          nebo absolutní počet bodů (tj. číslo mezi 1 a velikostí trénovací množiny), 
                          které budou pro každý podmodel rozhodovacího stromu náhodně vybrány z trénovací množiny (bootstrap) a použity k jeho trénování.
            max_depth - maximální hloubka každého z podmodelů rozhodovacího stromu.
            kwargs - (volitelně) případné další hyperparametry, které pošlete do podmodelů rozhodovacího stromu
        """
        self.n_estimators = n_estimators
        self.max_samples = max_samples 
        self.max_depth = max_depth
        self.trees = []
        
    def fit(self, X, y):
        """
        Natrénování modelu. Trénovací data jsou v argumentech X a y.
        Pro trénování podmodelů používejte bootstraping a velikost samplovaného vzorku vezměte z hyperparametru max_samples_fraction
        """
        for i in range(self.n_estimators):
            tree = DecisionTreeRegressor(max_depth=self.max_depth,)
            
            bootstrap = np.random.choice(X.index, size=int(X.shape[0]*self.max_samples), replace=True,)
            assert X.shape[0] == y.shape[0]
            
            Xsamples = X.loc[bootstrap]
            ysamples = y.loc[bootstrap]
            
            tree.fit(Xsamples, ysamples)
            self.trees.append(tree)

        
    def predict(self, X):
        """
        Predikce y v zadaných bodech X
        """
        pred_values = []
        for tree in self.trees:
            preds = np.array([tree.predict(X) for tree in self.trees])
            pred_values.append(tree.predict(X))

        ypredicted = np.mean(pred_values, axis=0)
        return ypredicted

        
                
        
````

````python
from tqdm import tqdm

df = pd.read_csv("data.csv")
param_grid = {
    'max_depth': range(5,15),
    "n_estimators": range(5,15),
    "max_samples": np.arange(0.1,1,0.1)
}

param_comb = ParameterGrid(param_grid)

RMSE_val_Forrest = []
RMSE_train_Forrest = []
for params in tqdm(param_comb):
    clf_forrest_screen = CustomRandomForest(max_depth=params['max_depth'], n_estimators=params['n_estimators'], max_samples = params['max_samples'], random_state=rd_seed, n_jobs=-1)
    clf_forrest_screen.fit(Xtrain, ytrain)
    RMSE_train_Forrest.append(np.sqrt(mean_squared_error(ytrain, clf_forrest_screen.predict(Xtrain))))
    RMSE_val_Forrest.append(np.sqrt(mean_squared_error(yval, clf_forrest_screen.predict(Xval))))

plt.figure(figsize=(20,6))
plt.plot(RMSE_train_Forrest,'or-')
plt.plot(RMSE_val_Forrest,'ob-')
plt.xlabel('hyperparametr index')
plt.ylabel('RMSE')
plt.legend(['train', 'validation'])

optimum_index = np.argmin(RMSE_val_Forrest)
print(f'Best parameters: {param_comb[optimum_index]},train RMSE: {RMSE_train_Forrest[optimum_index]}, val RMSE: {RMSE_val_Forrest[optimum_index]}')
````

````python
print(f'RMSE (val): {RMSE_val_Forrest[optimum_index]:.5f}')
MAE_val_Forrest = mean_absolute_error(yval, clf_forrest_screen.predict(Xval)) 
print(f'MAE (val): {MAE_val_Forrest:.5f}')
````

Jako nejlepší model náhodného lesa byl nalezen náhodný les s počtem podmodelů 11 s jejich maximální hloubkou 14 a pro bootstrap jsou z původní množiny vybírány náhodně body tak, aby množina podmodelu měla velikost 70 % původní množiny. Validační RMSE tohoto modelu vyšla 2.16199 a MAE 1.44006.
